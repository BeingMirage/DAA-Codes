```python
import sys

def matrix_chain_order(p):
    n = len(p)
    m = [[0 for _ in range(n)] for _ in range(n)]

    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost_left = m[i][k]              
                cost_right = m[k + 1][j]          
                cost_merge = p[i - 1] * p[k] * p[j] 

                total_cost = cost_left + cost_right + cost_merge

                if total_cost < m[i][j]:
                    m[i][j] = total_cost

    return m[1][n - 1]

arr = [10, 30, 5, 60]
print("Minimum number of multiplications:", matrix_chain_order(arr))
```
