```python
def insertion_sort(nums):
    for i in range(1, len(nums)):
        key = nums[i]
        j = i - 1

        # Shift elements greater than key one position to the right
        while j >= 0 and nums[j] > key:
            nums[j + 1] = nums[j]
            j -= 1

        nums[j + 1] = key  # Insert key in correct position

    print(nums)
```

```python
nums = [23, 44 ,34, 31, 72 ,3 ]
insertionsort(nums)
```
