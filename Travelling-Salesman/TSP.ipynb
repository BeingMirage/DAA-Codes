```python
import math

# Function to solve TSP using Backtracking
def tsp(graph, visited, curr_pos, n, count, cost, ans):
    # Base case: if all cities visited and can return to start
    if count == n and graph[curr_pos][0] > 0:
        ans[0] = min(ans[0], cost + graph[curr_pos][0])
        return

    # Try visiting all unvisited cities
    for city in range(n):
        if not visited[city] and graph[curr_pos][city] > 0:
            visited[city] = True
            tsp(graph, visited, city, n, count + 1, cost + graph[curr_pos][city], ans)
            visited[city] = False  # backtrack

# Example graph (adjacency matrix)
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

n = len(graph)
visited = [False] * n
visited[0] = True  # start from city 0
ans = [math.inf]

tsp(graph, visited, 0, n, 1, 0, ans)

print("Minimum cost to visit all cities:", ans[0])
```
