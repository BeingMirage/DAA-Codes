```python
N = 5  # board size

def print_board(board):
    for row in board:
        print(" ".join("Q" if c == 1 else "." for c in row))
    print()

def is_safe(board, row, col):
    # Check column and diagonals above the current row
    for i in range(row):
        if board[i][col] == 1 or \
           (col - (row - i) >= 0 and board[i][col - (row - i)] == 1) or \
           (col + (row - i) < N and board[i][col + (row - i)] == 1):
            return False
    return True

def solve(board, row=0):
    if row == N:
        print_board(board)
        return

    for col in range(N):
        if is_safe(board, row, col):
            board[row][col] = 1
            solve(board, row + 1)
            board[row][col] = 0  # backtrack

# Run program
solve([[0] * N for _ in range(N)])
```
