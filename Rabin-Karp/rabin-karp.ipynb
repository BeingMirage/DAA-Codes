```python
def rabin_karp(text, pattern, prime=101):
    n = len(text)
    m = len(pattern)
    d = 256  # number of possible characters (ASCII)
    p = 0    # hash value for pattern
    t = 0    # hash value for text window
    h = 1    # value of d^(m-1)

    # Precompute h = (d^(m-1)) % prime
    for _ in range(m - 1):
        h = (h * d) % prime

    # Calculate hash for pattern and first text window
    for i in range(m):
        p = (d * p + ord(pattern[i])) % prime
        t = (d * t + ord(text[i])) % prime

    # Slide pattern over text one by one
    for i in range(n - m + 1):
        # If hash matches, check characters one by one
        if p == t:
            if text[i:i+m] == pattern:
                print(f"Pattern found at index {i}")

        # Calculate hash for next window (remove first char, add next)
        if i < n - m:
            t = (d * (t - ord(text[i]) * h) + ord(text[i + m])) % prime

            # Ensure hash is positive
            if t < 0:
                t = t + prime


# Example
text = "AABAACAADAABAABA"
pattern = "AABA"

rabin_karp(text, pattern)
```
