```python
def activity_selection(start , finish):
       
       activities = sorted(zip(start,finish), key = lambda x:x[1])

       selected = []
       last_end = 0

       for s,f in activities:
              if s >= last_end:
                     selected.append((s,f))
                     last_end = f
       
       return selected
       

start = [1, 2, 4, 6, 5, 8]
finish = [3, 5, 6, 8, 9, 10]

result = activity_selection(start, finish)
print("Selected activities:", result)

```
