```python
def bubblesort(nums):
       for i in range(len(nums)):
              for j in range(i+1,len(nums)):
                     if nums[i] > nums[j]:
                            nums[i] , nums[j] = nums[j] , nums[i]
       
       print(nums)
```

```python
nums = [23, 44 ,34, 31, 72 ,3 ]
bubblesort(nums)
```
